```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
        
class DoublyLinkedList():    
    def __init__(self):
        self.head=None
    def InsertAtBeginning(self, data):
        if self.head == None:
            node = Node(data, self.head, None)
            self.head = node
        else:
            node = Node(data, self.head, None)
            self.head.prev = node
            self.head = node
            
    def print_forward(self):
        if self.head is None:
            print("Linked list is empty")
            return
        itr=self.head
        llstr=' '
        while itr:
            llstr +=str(itr.data)+'--->'
            itr= itr.next 
        print(llstr)  
    
    # TWO PASS APPROACH - TC- O(2N),  SC - O(N)
    
    def reverse_dll(self):
   
        if self.head is None or self.head.next is None:
            return  self.head

        # Initialize a stack to store values
        st = []
        temp = self.head
        while temp is not None:
            st.append(temp.data)
            temp = temp.next

        temp = self.head
        while temp is not None:
            temp.data = st.pop()
            # Traverse further
            temp = temp.next
        return self.head
    
    def reverse(self):
        if self.head is None or self.head.next is None:
            return self.head

        temp = self.head
        last = None  # To store the last valid node

        while temp:
            last = temp   # Save the last node
            temp.next, temp.prev = temp.prev, temp.next
            temp = temp.prev  # move to the next node (after swap)

        self.head = last
        return self.print_forward()
    
    def striver_reverse_dll(self):
    
        if self.head is None or self.head.next is None:
            return self.head
        
      
        prev = None  
        current = self.head  

        # Traverse the linked list
        while current is not None:
        
            prev = current.back 
            current.back = current.next
            current.next = prev 
            current = current.back  

        return prev.back

# Example usage:
arr = [12, 5, 6, 8, 4]
# Convert the array to a
# doubly linked list
                
```
