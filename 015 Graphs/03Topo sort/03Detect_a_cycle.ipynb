```python
from collections import deque
```

```python
def topo_sort(V,adj):
    indegree=[0]*V
    cnt =0
    for i in range(V):
        for neigbor in adj[i]:
            indegree[neigbor]+=1
            
    q = deque()  
    for i in range(V):
        if indegree[i]==0:
            q.append(i)

    while q:
        node = q.popleft()       
        cnt+=1      
        for neigbor in adj[node]:
            indegree[neigbor]-=1 
            if indegree[neigbor] == 0:
                q.append(neigbor)
    if cnt ==V:
        return False        
    return True     
V = 6
# Adjacency list representation matching your C++ example:
adj = [
    [],         # 0
    [],         # 1
    [3],        # 2
    [1],        # 3
    [0, 1],     # 4
    [0, 2]      # 5
]

result = topo_sort(V, adj)
print("Topological order:", result)
```
