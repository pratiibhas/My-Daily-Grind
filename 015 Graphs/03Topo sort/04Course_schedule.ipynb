```python
from collections import deque
class Solution:
    def canFinish(self, numCourses: int, prerequisites) :
        cnt =0
        indegree = [0]*numCourses
        adj = [[] for _ in range(numCourses)]  # adjacency list
        # Build graph and indegree
        for dest, src in prerequisites:
            adj[src].append(dest)  
            indegree[dest] += 1
        q = deque()
        for i in range(numCourses):
            if indegree[i]==0:
                q.append(i)
        while q:
            node = q.popleft()
            cnt+=1
            for neigbor in adj[node]:
                indegree[neigbor]-=1
                if indegree[neigbor]==0:
                    q.append(neigbor)
        if cnt == numCourses:
            return True
        return False                  

        
```

```python
numCourses = 2
prerequisites = [[1,0]]
Solution().canFinish(numCourses,prerequisites)
```
