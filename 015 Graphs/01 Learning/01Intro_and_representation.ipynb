## Representation 
- n- nodes , m - edges
- two ways .a. Adjacency Matrix b. list

### Adjacency matrix

```python
def main():
    n, m = map(int, input().split())
    
    # adjacency matrix for undirected graph
    # initializing with 0
    adj = [[0] * (n + 1) for _ in range(n + 1)]
    
    for _ in range(m):
        u, v = map(int, input().split())
        adj[u][v] = 1
        adj[v][u] = 1  # remove this line for a directed graph
    
    # Just showing adjacency matrix (optional)
    for row in adj:
        print(row)


if __name__ == "__main__":
    main()
```

### Adjacency list

### For undirected graph

```python
def main():
    n, m = map(int, input().split())
    
   
    # initializing with 0
    adj = [[] for _ in range(n + 1)]
    for i in range(m):
        u, v = map(int, input().split())
        adj[u].append(v)
        adj[v].append(u)
    for row in adj:
        print(row)    
    return 0
if __name__ == "__main__":
    main()     
```

SC- O(2*E)

### For Directed graph

```python
def main():
    n, m = map(int, input().split())
   
    # initializing with 0
    adj = [[] for _ in range(n + 1)]
    for _ in range(m):
        u, v = map(int, input().split())
        adj[u].append(v)
    for row in adj:
        print(row)    
    return 0
if __name__ == "__main__":
    main() 
```

SC- O(E)

## For weighted graphs

### Adjacency matrix

```python
def main():
    n, m = map(int, input().split())
    
    # adjacency matrix for undirected graph
    # initializing with 0
    adj = [[0] * (n + 1) for _ in range(n + 1)]
    
    for _ in range(m):
        u, v ,w = map(int, input().split())
        adj[u][v] = w
        adj[v][u] = w  # remove this line for a directed graph
    
    # Just showing adjacency matrix (optional)
    for row in adj:
        print(row)


if __name__ == "__main__":
    main()
```

### Adjacency list (undirected)

```python
def main():
    n, m = map(int, input().split())
    
    # adjacency list for weighted undirected graph
    # each element is a list of (neighbor, weight)
    adj = [[] for _ in range(n + 1)]
    
    for _ in range(m):
        u, v, w = map(int, input().split())
        adj[u].append((v, w))
        adj[v].append((u, w))  # remove this line for directed graph
    
    # Example printout (optional)
    for i in range(1, n + 1):
        print(f"Node {i}: {adj[i]}")

if __name__ == "__main__":
    main()
```

### Adjacency list (directed)

```python
def main():
    n, m = map(int, input().split())
    
    # adjacency list for weighted directed graph
    # each element is a list of (neighbor, weight)
    adj = [[] for _ in range(n + 1)]
    
    for _ in range(m):
        u, v, w = map(int, input().split())
        adj[u].append((v, w))
       
    
    # Example printout (optional)
    for i in range(1, n + 1):
        print(f"Node {i}: {adj[i]}")

if __name__ == "__main__":
    main()
```
