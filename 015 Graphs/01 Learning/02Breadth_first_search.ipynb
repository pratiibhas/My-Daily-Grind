## Breadth First s earch

```python
from collections import deque

class Solution:
    def bfsOfGraph(self, V, adj):
        vis = [False] * V
        vis[0] = True
        q = deque([0])
        bfs = []
        
        while q:
            node = q.popleft()
            bfs.append(node)
            for neighbor in adj[node]:
                if not vis[neighbor]:
                    vis[neighbor] = True
                    q.append(neighbor)
        return bfs

def addEdge(adj, u, v):
    adj[u].append(v)
    adj[v].append(u)

def printAns(ans):
    print(' '.join(map(str, ans)))

def main():
    V = 6
    adj = [[] for _ in range(V)]
    
    addEdge(adj, 0, 1)
    addEdge(adj, 1, 2)
    addEdge(adj, 1, 3)
    addEdge(adj, 0, 4)
    
    obj = Solution()
    ans = obj.bfsOfGraph(5, adj)
    print(ans)
    printAns(ans) # graph version

if __name__ == "__main__":
    main()
```
