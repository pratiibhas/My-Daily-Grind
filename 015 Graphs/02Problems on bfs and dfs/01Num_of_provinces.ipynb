```python
from collections import deque
from typing import List

class Solution:
    def dfs(self,node,adj,visited):    
        visited[node] = True

        
        for neigbor in adj[node]:
            if not visited[neigbor]:
                self.dfs(neigbor,adj,visited)
              

    def numProvinces(self, adj: List[List[int]], V: int) -> int:
        # Convert adjacency matrix to adjacency list
        adjls = [[] for _ in range(V)]
        for i in range(V):
            for j in range(V):
                if adj[i][j] == 1 and i != j:
                    adjls[i].append(j)
                    adjls[j].append(i)
        vis = [0] * V
        cnt = 0
        for i in range(V):
            if not vis[i]:
                cnt += 1
                self.dfs(i, adjls, vis)
        return cnt
        
    
```

```python
isConnected = [[1,1,0],[1,1,0],[0,0,1]]
Solution().numProvinces(isConnected,3)
```
