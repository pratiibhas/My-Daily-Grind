## Surrounded regions
You are given an m x n matrix board containing letters 'X' and 'O', capture regions that are surrounded:

- Connect: A cell is connected to adjacent cells horizontally or vertically.
- Region: To form a region connect every 'O' cell.
- Surround: The region is surrounded with 'X' cells if you can connect the region with 'X' cells and none of the region cells are on the edge of the board.
To capture a surrounded region, replace all 'O's with 'X's in-place within the original board. You do not need to return anything.

 

Example 1:

Input: board = [["X","X","X","X"],["X","O","O","X"],["X","X","O","X"],["X","O","X","X"]]

Output: [["X","X","X","X"],["X","X","X","X"],["X","X","X","X"],["X","O","X","X"]]

- If on boundary cannot be converted to X
- IF a set of O's connected have a O on boundary cannot be converted to X.

```python
class Solution:
    def __init__(self):
        # Directions for top, right, bottom, left
        self.delrow = [-1, 0, 1, 0]
        self.delcol = [0, 1, 0, -1]

    def dfs(self, row, col, vis, mat, n, m):
        vis[row][col] = 1
        for i in range(4):
            nrow = row + self.delrow[i]
            ncol = col + self.delcol[i]
            # Check within bounds and unvisited 'O'
            if (0 <= nrow < n and 0 <= ncol < m 
                and not vis[nrow][ncol] 
                and mat[nrow][ncol] == 'O'):
                self.dfs(nrow, ncol, vis, mat, n, m)

    def fill(self, n, m, mat):
        vis = [[0 for _ in range(m)] for _ in range(n)]

        # First and last row
        for j in range(m):
            if mat[0][j] == 'O' and not vis[0][j]:
                self.dfs(0, j, vis, mat, n, m)
            if mat[n-1][j] == 'O' and not vis[n-1][j]:
                self.dfs(n-1, j, vis, mat, n, m)

        # First and last column
        for i in range(n):
            if mat[i][0] == 'O' and not vis[i][0]:
                self.dfs(i, 0, vis, mat, n, m)
            if mat[i][m-1] == 'O' and not vis[i][m-1]:
                self.dfs(i, m-1, vis, mat, n, m)

        # Flip unvisited 'O' to 'X'
        for i in range(n):
            for j in range(m):
                if mat[i][j] == 'O' and not vis[i][j]:
                    mat[i][j] = 'X'
        return mat


# Example run
if __name__ == "__main__":
    mat = [
        ['X', 'X', 'X', 'X'],
        ['X', 'O', 'X', 'X'],
        ['X', 'O', 'O', 'X'],
        ['X', 'O', 'X', 'X'],
        ['X', 'X', 'O', 'O']
    ]

    ob = Solution()
    ans = ob.fill(5, 4, mat)
    for row in ans:
        print(' '.join(row))
```
