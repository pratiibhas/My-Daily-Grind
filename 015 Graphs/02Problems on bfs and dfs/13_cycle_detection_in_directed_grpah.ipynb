##  Detect cycle in a directed graph using DFS

```python
def dfs(node,adj,vis,path):
    vis[node] = 1
    path[node] = 1
    for v in adj[node]:
        if not vis[v]:
            if dfs(v,adj,vis,path):
                return True
            # If the node has been previously visited and it has to be visited on the same path
        elif path[v]:
            return True
    path[node]=0
    return False   
                
def cycle_exists(graph):
    n = len(graph)
    vis = [0]*n
    path_visited =[0]*n
    
    for i in range(n):
        if not vis[i]:
            if dfs(i,graph,vis,path_visited):
                return True
    return False            
        
```

```python
adj = [
    [1],   # 0 → 1
    [2],   # 1 → 2
    [0,3], # 2 → 0 (cycle), 2 → 3
    []     # 3
]
cycle_exists(adj)
```

```python
adj =[[1], [2], [], [0,2]]
cycle_exists(adj)
```
