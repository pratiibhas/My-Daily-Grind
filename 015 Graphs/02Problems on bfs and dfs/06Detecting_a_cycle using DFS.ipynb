## Detecting a cycle using DFS

```python
from collections import deque
from typing import List

class Solution:
    def dfs(self, node: int,parent, adj: List[List[int]], vis: List[int]) -> bool:
        vis[node] = 1
        
        for adjacent_node in adj[node]:
            if not vis[adjacent_node]:
                if self.dfs(adjacent_node,node,adj,vis)== True:
                    return True
            # if adjacent node is visited and is not it's own parent node
            elif adjacent_node != parent:
                return True  # Cycle detected
        return False

    def isCycle(self, V: int, adj: List[List[int]]) -> bool:
        vis = [0] * V
        # in case disconnected components exists, for other cases if we give a starting node it will work alright 
        for i in range(V):
            if not vis[i]:
                if self.dfs(i,-1, adj, vis)== True:
                    return True
        return False

if __name__ == "__main__":
    # Example graph with 4 vertices
    adj = [
        [],     # Node 0 has no adjacent nodes
        [2],    # Node 1 adjacent to 2
        [1, 3], # Node 2 adjacent to 1 and 3
        [2]     # Node 3 adjacent to 2
    ]

    obj = Solution()
    ans = obj.isCycle(4, adj)
    print(1 if ans else 0)
```

SC- O(N)
TC- O(N+2E) + O(N)
