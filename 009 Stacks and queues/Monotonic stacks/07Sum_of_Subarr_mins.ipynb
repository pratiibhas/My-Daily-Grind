## Sum of Subarray Minimums

Given an array of integers arr, find the sum of min(b), where b ranges over every (contiguous) subarray of arr. Since the answer may be large, return the answer modulo 109 + 7.

 

Example 1:

Input: arr = [3,1,2,4]
Output: 17
Explanation: 
Subarrays are [3], [1], [2], [4], [3,1], [1,2], [2,4], [3,1,2], [1,2,4], [3,1,2,4]. 
Minimums are 3, 1, 2, 4, 1, 1, 2, 1, 1, 1.
Sum is 17.
Example 2:

Input: arr = [11,81,94,43,3]
Output: 444


```python
from typing import List

class Solution:
    def find_nse(self, arr: List[int]) -> List[int]:
        n = len(arr)
        stack = []
        res = [n] * n  # default is n (means no smaller to the right)
        for i in range(n - 1, -1, -1):
            while stack and arr[stack[-1]] >= arr[i]:
                stack.pop()
            if stack:
                res[i] = stack[-1]
            stack.append(i)
        return res

    def find_pse(self, arr: List[int]) -> List[int]:
        n = len(arr)
        stack = []
        res = [-1] * n  # default is -1 (means no smaller to the left)
        for i in range(n):
            while stack and arr[stack[-1]] > arr[i]:
                stack.pop()
            if stack:
                res[i] = stack[-1]
            stack.append(i)
        return res

    def sumSubarrayMins(self, arr: List[int]) -> int:
        mod = 10**9 + 7
        n = len(arr)
        nse = self.find_nse(arr)
        pse = self.find_pse(arr)
        
        total = 0
        for i in range(n):
            left = i - pse[i]
            right = nse[i] - i
            total = (total + arr[i] * left * right) % mod
        return total
```
