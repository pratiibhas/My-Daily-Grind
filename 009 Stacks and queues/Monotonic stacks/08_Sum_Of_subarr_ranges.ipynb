## Sum of subarray ranges

### Brute force

```python
def subArrayRanges(nums) -> int:
    sumi =0
    n = len(nums)
    for i in range(n):
        largest = nums[i]
        smallest= nums[i]
        for j in range(i+1,n):
            largest = max(largest, nums[j])
            smallest = min(smallest,nums[j])

            sumi+= (largest-smallest)
    return sumi   
```

```python
nums =[3,4,5,7,2,1,4]
subArrayRanges(nums)
```

### Optimal version

```python
from typing import List


def find_nse(arr: List[int]) -> List[int]:
    n = len(arr)
    stack = []
    res = [n] * n  # default is n (means no smaller to the right)
    # for handling the edge cases ( when there is no next smaller element)
    for i in range(n - 1, -1, -1):
        while stack and arr[stack[-1]] >= arr[i]:
            stack.pop()
        if stack:
            res[i] = stack[-1]
        stack.append(i)
    return res

def find_pse(arr: List[int]) -> List[int]:
    n = len(arr)
    stack = []
    res = [-1] * n  # default is -1 (means no smaller to the left)
    # for handling the edge cases ( when there is no prev smaller element)
    for i in range(n):
        while stack and arr[stack[-1]] > arr[i]:
            stack.pop()
        if stack:
            res[i] = stack[-1]
        stack.append(i)
    return res

def sumSubarrayMins(arr: List[int]) -> int:
    mod = 10**9 + 7
    n = len(arr)
    nse = find_nse(arr)
    pse = find_pse(arr)
    
    total = 0
    for i in range(n):
        left = i - pse[i]
        right = nse[i] - i
        total = (total + arr[i] * left * right) % mod
    return total
def find_nge(arr):
    n = len(arr)
    nge =[n]*n
    st =[]
    
    for i in range(n-1,-1,-1):
        while st and arr[st[-1]] <= arr[i]:
            st.pop()
        
        if not st:
            nge[i]=n
        else:
            nge[i] = st[-1]
        st.append(i)
    return nge 

def find_pge(arr):
    n = len(arr)
    nge =[-1]*n
    st =[]
    
    for i in range(n):
        while st and arr[st[-1]] < arr[i]:
            st.pop()
        
        if not st:
            nge[i]=-1
        else:
            nge[i] = st[-1]
        st.append(i)
    return nge 

def sumSubarrayMax(arr: List[int]) -> int:
    mod = 10**9 + 7
    n = len(arr)
    nge = find_nge(arr)
    pge = find_pge(arr)
    
    total = 0
    for i in range(n):
        left = i - pge[i]
        right = nge[i] - i
        total = (total + arr[i] * left * right) % mod
    return total
```

```python
def sumofranges(arr):
    return sumSubarrayMax(arr)- sumSubarrayMins(arr)
```

```python
arr=[1,4,3,2]
sumofranges(arr)
```
