```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Queue:

    # Initializing a stack.
    # Use a dummy node, which is
    # easier for handling edge cases.
    def __init__(self):
        self.head = Node("head")
        self.size = 0
        self.start=self.head
        self.end= self.head
        
    def push(self,x):
        node=Node(x)
        if self.size==0:
           self.start = node
        self.end.next=node
        self.end = node        # Update the end pointer
        self.size += 1 
        
    def pop(self):
        if(self.start is None):
            return
        ele=self.start.value
        self.start=self.start.next 
        return ele
    
    def top(self):
        return self.start.value
        
    def __str__(self):
        if self.size == 0:
            return "Queue is empty"
        elements = []
        current = self.start
        while current:
            elements.append(str(current.value))
            current = current.next
        return "->".join(elements) 
      
if __name__ == "__main__":
    q = Queue()
    q.push(10)
    q.push(20)
    q.push(30)
    print(q)  
    print(q.pop())  # Output: 10
    print(q)  # Output: 20->30

    q.push(40)
    q.push(50)
    q.push(60)
    print(q)   
    print(q.top())     
```

### Implementation of Queues using stack

```python
from collections import deque

class Queue:
    def __init__(self):
        self.size = 0
        self.stack = deque()   # Primary stack
        self.stack2 = deque()  # Temporary stack for rearranging elements

    def push(self, x):   
        # If the stack is empty, just add the element
        if self.size == 0:
            self.stack.append(x)
        else:
            # Move all elements from stack to stack2
            while self.stack:
                self.stack2.append(self.stack.popleft())
            # Add the new element to stack
            self.stack.append(x)
            # Move everything back from stack2 to stack
            while self.stack2:
                self.stack.append(self.stack2.popleft())
        # Increment the size of the queue
        self.size += 1

    def pop(self):
        if not self.stack2:
            # Move all elements from stack1 to stack2 if stack2 is empty
            while self.stack:
                self.stack2.append(self.stack.pop())
        if not self.stack2:
            print("Queue is empty!")
            return None
        # Pop the front element of the queue (top of stack2)
        return self.stack2.pop()
    
    def peek(self):
        if self.size == 0:
            print("Queue is empty!")
            return None
        # Return the front element without removing it
        return self.stack[0]

    def is_empty(self):
        return self.size == 0

    def __str__(self):
        return f"Queue contents: {list(self.stack)}"

# Example usage:
q = Queue()
q.push(10)
q.push(20)
q.push(30)

print(q)  # Output: Queue contents: [10, 20, 30]
print(q.pop())  # Output: 10
print(q)  # Output: Queue contents: [20, 30]
print(q.peek())  # Output: 20
```
